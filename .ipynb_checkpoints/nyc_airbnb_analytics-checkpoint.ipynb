```python
import pandas as pd

# File path to the dataset
file_path = r"D:\NYC_AirBnb_analytics\listings.csv"

# Load the dataset
data = pd.read_csv(file_path)

# Display basic information
print("Dataset loaded successfully!")
print(f"Dataset contains {data.shape[0]} rows and {data.shape[1]} columns.")
print(data.head())
```

```python
print(data.info())
print(data.isnull().sum())
```

```python
data = data.copy()  # Ensure you are working with a copy of the original DataFrame

# Handle missing values
data['price'] = data['price'].fillna(data['price'].median())
data['reviews_per_month'] = data['reviews_per_month'].fillna(0)
data['name'] = data['name'].fillna('Unknown')
data['host_name'] = data['host_name'].fillna('Unknown')

# Convert last_review to datetime
data['last_review'] = pd.to_datetime(data['last_review'], errors='coerce')

print(data.isnull().sum())
print(data.info())
```

```python
# Replace positive and negative infinity with max and min finite values
X.replace([float('inf'), -float('inf')], X.max().max(), inplace=True)
print(f"Any infinite values left: {X.isin([float('inf'), -float('inf')]).values.any()}")
for col in X.columns:
    if X[col].isin([float('inf'), -float('inf')]).any():
        print(f"Infinite values found in column: {col}")
for col in X.columns:
    if X[col].isin([float('inf'), -float('inf')]).any():
        max_finite_value = X.loc[X[col] != float('inf'), col].max()  # Find the maximum finite value
        min_finite_value = X.loc[X[col] != -float('inf'), col].min()  # Find the minimum finite value
        X[col].replace([float('inf')], max_finite_value, inplace=True)
        X[col].replace([-float('inf')], min_finite_value, inplace=True)
print(f"Any infinite values left: {X.isin([float('inf'), -float('inf')]).values.any()}")
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize scaler
scaler = StandardScaler()

# Scale the dataset
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error, r2_score

# Initialize Ridge Regression
ridge = Ridge(alpha=1.0)

# Train the model
ridge.fit(X_train, y_train)

# Make predictions
y_pred = ridge.predict(X_test)

# Evaluate the model
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error: {mae}")
print(f"R Score: {r2}")
```

```python
from sklearn.model_selection import GridSearchCV

# Define parameter grid
param_grid = {'alpha': [0.1, 1, 10, 100, 1000]}

# Perform grid search
ridge_grid = GridSearchCV(Ridge(), param_grid, cv=5, scoring='r2')
ridge_grid.fit(X_train, y_train)

# Best parameters and score
print(f"Best Alpha: {ridge_grid.best_params_['alpha']}")
print(f"Best R Score: {ridge_grid.best_score_}")

```
